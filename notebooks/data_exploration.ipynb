```python
import os
```

```python
# Read a file from the data/transcripts directory which is one level up from the current directory
def read_file(filename):
    with open(os.path.join('..', 'data', 'transcripts', filename), 'r') as f:
        return f.read()
```

```python
print(read_file('episode1'))
```

```python
print(read_file('episode2'))
```

```python
print(read_file('episode4'))
```
